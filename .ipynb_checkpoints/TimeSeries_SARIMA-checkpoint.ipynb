# STL Decomposition of stock price Time Series

```python
ms_res = sts.seasonal_decompose(ms['Adj Close'], model = 'additive',freq=52)
#I have used 52 in the freq parameter. I am not entirely sure if thats appropriate.
```

```python
ms_res.plot()
plt.show()
```

Results:

1. The periodic component looks reasonable. 
2. The removal of the trend component appears to be satisfactory.

We can apply the **Dicky Fuller** test to determine if the residual is stationary. The null hypothesis is that the time series is not stationary, but with no trend.

```python
ms_res.resid[26:-26]
```

```python
def DF_Test(ts):
    stationary = adfuller(ts)
    ## Print the results
    print('D-F statistic = ' + str(stationary[0]))
    print('p-value = ' + str(stationary[1]))
    print('number of lags used = ' + str(stationary[2]))
    print('Critical value at 5% confidence = ' + str(stationary[4]['5%']))
    print('Critical value at 10% confidence = ' + str(stationary[4]['10%']))

DF_Test(ms_res.resid[26:-26])
```

Given the DF statistic and p-value we can **reject** the null hypothesis that the residual is not stationary.

As a next step, compute and plot the ACF and PACF for the remainder series.

```python
splt.plot_acf(ms_res.resid[26:-26], lags = 60)
splt.plot_pacf(ms_res.resid[26:-26], lags = 60)
plt.show()
```

Results from ACF and PACF plots
1. ACF shows significant MA behaviour upto lags 30
2. PACF shows AR behaviour for lags 2,5

### Find the best fit ARIMA Model

```python
def model_ARIMA(ts, order):
    from statsmodels.tsa.arima_model import ARIMA
    model = ARIMA(ts, order = order)
    model_fit = model.fit(disp=0, method='mle', trend='nc')
    print(model_fit.summary())
    return(model_fit)

def model_ARIMA_2(ts, order):
    from statsmodels.tsa.arima_model import ARIMA
    from statsmodels.tsa.arima_model import ARIMAResults    
    model = ARIMA(ts, order = order)
    model_fit = model.fit(disp=0, method='mle', trend='nc')
    BIC = ARIMAResults.bic(model_fit)
    print('Testing model of order: ' + str(order) + ' with BIC = ' + str(BIC))
    return(BIC, order, model_fit)

def step_ARIMA(resid, p_max, d_max, q_max):
    from statsmodels.tsa.arima_model import ARIMAResults    
    from statsmodels.tsa.arima_model import ARIMA
    best_BIC = 9999999999999999.0
    for p in range(p_max + 1):
        for d in range(d_max + 1):
            for q in range(q_max + 1):
                if(p > 0 or q > 0):
                    try:
                        order = (p, d, q)
                        BIC, order, model = model_ARIMA_2(resid, order)
                        if(BIC < best_BIC):
                            best_model = model
                            best_BIC = BIC
                            best_order = order
                    except: 
                        pass        
    return(best_BIC, best_order, best_model) 

BIC, order, model = step_ARIMA(ms_res.resid[26:-26], 2, 2, 2)    
print('***************************************')
print('Best model with BIC = ' + str(BIC) + ' and with order '+ str(order))
```

# MSFT stock price forecasting 

```python
#create new ms stock dataset
new_ms = pd.DataFrame(index=range(0,len(ms)),columns=['Date', 'close'])
for i in range(0,len(ms)):
    new_ms['Date'][i] = ms['Date'][i]
    new_ms['close'][i] = ms['Adj Close'][i]
new_ms.head(10)
```

```python
#create more features such as ‘Year’, ‘Month’, ‘Week’, ‘Day’, ‘Dayofweek’, ‘Dayofyear’, 
#‘Is_month_end’, ‘Is_month_start’, ‘Is_quarter_end’, ‘Is_quarter_start’,  ‘Is_year_end’, and  ‘Is_year_start’.

new_ms['year'] = new_ms['Date'].dt.year
new_ms['month'] = new_ms['Date'].dt.month
new_ms['day'] = new_ms['Date'].dt.day
new_ms['dayofweek'] = new_ms['Date'].dt.dayofweek
new_ms['weekofyear'] = new_ms['Date'].dt.weekofyear
new_ms['dayofyear'] = new_ms['Date'].dt.dayofyear
new_ms['is_month_end'] = new_ms['Date'].dt.is_month_end.astype(int)
new_ms['is_month_start'] = new_ms['Date'].dt.is_month_start.astype(int)
new_ms['is_quarter_end'] = new_ms['Date'].dt.is_quarter_end.astype(int)
new_ms['is_quarter_start'] = new_ms['Date'].dt.is_quarter_start.astype(int)
new_ms['is_year_end'] = new_ms['Date'].dt.is_year_end.astype(int)
new_ms.head(10)
```

```python
#We will now split the data into train and validation sets to check the performance of the model.
train = new_ms[:629]
valid = new_ms[630:]
```

```python
x_train = train.drop('close', axis=1)
x_train = train.drop('Date', axis=1)
y_train = train['close']
x_valid = valid.drop('close', axis=1)
x_valid = valid.drop('Date', axis=1)
y_valid = valid['close']

#implement linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#make predictions and find the rmse
preds = model.predict(x_valid)
rms=np.sqrt(np.mean(np.power((np.array(y_valid)-np.array(preds)),2)))
rms
```

```python
#plot
valid['Predictions'] = 0
valid['Predictions'] = preds

#ms = ms.set_index(pd.to_datetime(ms['Date'], format= '%m/%d/%Y'))
valid = valid.set_index(new_ms[630:].index)
train = train.set_index(new_ms[:629].index)

plt.plot(train['close'],color = 'b')
plt.plot(valid['Predictions'],color='r',linewidth=1.0)
plt.plot(valid['close'],color = 'g',linewidth=5.0,alpha=0.5)
plt.show()
```
